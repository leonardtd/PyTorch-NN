# 4. Content, style features, gram matrix

**Paper:** </b>
- Style Features: conv1_1, conv2_1, conv3_1, conv4_1, conv5_1
- Content Features: conv4_2

```python
content_f = utils.get_features(content_p, vgg)
style_f = utils.get_features(style_p, vgg)
```

### Gram matrix

```python
style_grams = {layer: utils.gram_matrix(style_f[layer]) for layer in style_f}
```

# 5. Content & Style Loss

```python
### weighted mean
style_weights = {
    'conv1_1': 1,
    'conv2_1': 0.75,
    'conv3_1': 0.2,
    'conv4_1': 0.2,
    'conv5_1': 0.2,
}
```

```python
### TARGET: can be initialized with random noise or with content image
### this will be trained to minimize combination of losses


target = content_p.clone().requires_grad_(True).to(device)
#target = torch.randn_like(content_p).requires_grad_(True).to(device)
```

```python
target_features = utils.get_features(target, vgg)

print("Content Loss: ", utils.content_loss(target_features['conv4_2'], content_f['conv4_2']))
print("Style Loss: ", utils.style_loss(style_weights, target_features, style_grams))
```

# 6. Training Loop

```python
from torch import optim
```

```python
### HYPERPARAMS

optimizer = optim.Adam([target], lr=1e-3)

alpha = 1
beta = 1e5

epochs = 3000
show_every = 500
```

```python
results = []

for i in range(epochs):
    target_features = utils.get_features(target, vgg)
    
    c_loss = utils.content_loss(target_features['conv4_2'], content_f['conv4_2'])
    s_loss = utils.style_loss(style_weights, target_features, style_grams)
    t_loss = utils.total_loss(c_loss, s_loss, alpha, beta)
    
    optimizer.zero_grad()
    t_loss.backward()
    optimizer.step()
    
    if i% show_every == 0:
        print("Total Loss at epoch {} : {}".format(i, t_loss.item()))
        results.append(utils.deprocess(target.detach()))
```

```python
plt.figure(figsize=(10,8))

for i in range(len(results)):
    plt.subplot(3,2,i+1)
    plt.imshow(results[i])
plt.show()
```
